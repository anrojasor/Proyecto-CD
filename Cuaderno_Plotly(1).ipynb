## <span style="color:#4361EE">Introducción a Plotly</span>

Plotly es una librería de visualización de datos interactiva y de código abierto en Python que permite crear gráficos y visualizaciones de alta calidad. Fue desarrollada para facilitar la tarea de explorar, analizar y presentar datos de manera efectiva. Con Plotly, los científicos de datos, analistas y programadores pueden crear visualizaciones atractivas e interactivas con facilidad.

### <span style="color:#4CC9F0"> Principales características de Plotly</span>

*Interactividad*: Una de las características más destacadas de Plotly es su capacidad para crear gráficos interactivos. Los gráficos generados por Plotly permiten a los usuarios explorar y manipular los datos directamente en el gráfico. Puedes hacer zoom, desplazarte, seleccionar puntos específicos y obtener información detallada al colocar el cursor sobre los elementos.

*Variedad de gráficos*: Plotly ofrece una amplia variedad de tipos de gráficos, desde gráficos de líneas y gráficos de dispersión hasta gráficos de barras, histogramas, gráficos de pastel y mapas de calor. Esto permite representar tus datos de diferentes maneras y elegir el tipo de gráfico más adecuado para resaltar patrones y relaciones en los datos.

*Facilidad de uso*: Plotly se destaca por su facilidad de uso y sintaxis sencilla. La librería está diseñada para ser intuitiva y accesible para aquellos que deseen crear visualizaciones sin una curva de aprendizaje compleja.

*Visualización enriquecida*: Los gráficos generados por Plotly son altamente personalizables. Puedes ajustar aspectos como colores, títulos, etiquetas de ejes y leyendas para mejorar la presentación visual de tus datos.

*Integración con entornos de datos*: Plotly se integra perfectamente con Jupyter Notebooks, lo que facilita la creación de visualizaciones interactivas junto con el análisis de datos y la narrativa de proyectos.

*Soporte multiplataforma*: Además de Python, Plotly está disponible en otros lenguajes de programación como R, JavaScript y Julia, lo que facilita su adopción en diferentes entornos y flujos de trabajo.

### <span style="color:#4CC9F0">Aplicaciones de Plotly en Ciencia de datos</span>

*Exploración de datos*: Plotly es una herramienta poderosa para explorar datos y detectar patrones, tendencias y relaciones en los datos de manera interactiva.

*Presentación de resultados*: Con Plotly, puedes crear visualizaciones atractivas y dinámicas para presentar tus resultados y hallazgos de manera clara y convincente.

*Visualización geoespacial*: Plotly permite crear mapas interactivos y mapas cloropléticos para representar datos geoespaciales y resaltar patrones regionales.

*Visualización temporal*: Puedes utilizar Plotly para mostrar cambios en los datos a lo largo del tiempo mediante animaciones y gráficos interactivos.

## <span style="color:#4361EE">Introducción a Plotly express</span>

Plotly Express es una librería de Python desarrollada por Plotly que proporciona una interfaz de alto nivel para la creación rápida y sencilla de visualizaciones interactivas. Diseñada para simplificar el proceso de generación de gráficos, Plotly Express se ha convertido en una herramienta poderosa para científicos de datos, analistas y desarrolladores que buscan explorar y comunicar sus datos de manera efectiva.

Con Plotly Express, no es necesario sumergirse en complejas configuraciones o sintaxis detallada. En lugar de eso, puedes crear visualizaciones impresionantes con solo unas pocas líneas de código. La biblioteca ofrece una amplia variedad de tipos de gráficos, desde gráficos de barras, gráficos de líneas y gráficos de dispersión, hasta gráficos de pastel, histogramas, mapas de calor y más.

### <span style="color:#4361EE">Gráficos básicos</span>

#### <span style="color:#4361EE">Scatter</span>

En el primer gráfico, utilizamos el conjunto de datos "gapminder" y filtramos los datos para que solo incluyan el año 2007. Luego, creamos un diagrama de dispersión (scatter plot) utilizando Plotly Express. El eje x representa el PIB per cápita, mientras que el eje y representa la esperanza de vida. Los puntos se agrupan por continente y se les asigna un color único para cada continente. Además, al pasar el cursor sobre un punto, se mostrará el nombre del país. Este gráfico nos permite visualizar la relación entre el PIB per cápita, la esperanza de vida y el continente al que pertenece cada país en el año 2007.

````python
df = px.data.gapminder().query("year == 2007")
fig = px.scatter(df, x="gdpPercap", y="lifeExp", color="continent", hover_name="country")
fig.show()
````

En el segundo gráfico, creamos otro diagrama de dispersión, pero esta vez ajustamos el tamaño de los puntos según la población y aplicamos una escala logarítmica al eje x. De esta manera, los países con mayor población tendrán puntos más grandes. Los colores siguen representando los continentes y al pasar el cursor sobre un punto, se mostrará el nombre del país. Este gráfico nos permite comparar la relación entre el PIB per cápita, la esperanza de vida y la población de cada país en el año 2007, teniendo en cuenta la distribución logarítmica del PIB per cápita.

````python
df = px.data.gapminder().query("year == 2007")
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent", hover_name="country", log_x=True)
fig.show()
````

En el tercer gráfico, creamos un diagrama de dispersión similar al segundo gráfico, pero esta vez agregamos una línea de tendencia lineal mediante regresión por mínimos cuadrados. La línea de tendencia muestra la relación general entre el PIB per cápita y la esperanza de vida para los países en el año 2007. Los puntos todavía se agrupan por continente y tienen un tamaño proporcional a la población. Al pasar el cursor sobre un punto, se mostrará el nombre del país. La línea de tendencia "ols" (regresión por mínimos cuadrados) nos ayuda a identificar la tendencia general de cómo el PIB per cápita se relaciona con la esperanza de vida en ese año.

````python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent", hover_name="country", trendline="ols", log_x=True)
fig.show()
````

En el cuarto gráfico, agregamos gráficos marginales en los ejes x e y en forma de violín y caja, respectivamente. Estos gráficos marginales nos permiten explorar la distribución de las variables individuales en los ejes x e y mientras observamos el diagrama de dispersión principal. Además, mantenemos la línea de tendencia "ols" para ver la tendencia general entre el PIB per cápita y la esperanza de vida. Al pasar el cursor sobre un punto, se mostrará el nombre del país. Este gráfico ofrece una visión más detallada de cómo se distribuyen las variables y cómo se relacionan entre sí en el año 2007.

````python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent", hover_name="country", marginal_y="violin", marginal_x="box", trendline="ols", log_x=True)
fig.show()
````

````python
fig = px.scatter_matrix(df, dimensions=["lifeExp", "pop", "gdpPercap"], color="continent")
fig.show()
````

````python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country", facet_col="continent",
           log_x=True, size_max=45, range_x=[100,100000], range_y=[25,90])
fig.show()
````

#### <span style="color:#4361EE">Bar Plot</span>

Utilizamos px.bar() para crear el gráfico de barras. Pasamos el DataFrame df como el primer argumento. Especificamos que el eje x corresponde a la columna "continent", el eje y corresponde a la columna "pop" (población) y el color de las barras se basará en la columna "continent". También agregamos un título al gráfico utilizando el parámetro title, y utilizamos el parámetro labels para cambiar los nombres de las etiquetas de los ejes y hacer el gráfico más descriptivo.

El resultado será un gráfico de barras que muestra la población por continente en el año 2007. Cada barra representa la población de un continente específico, y los colores de las barras representan los diferentes continentes.

````python
# Crear el gráfico de barras con Plotly Express
fig = px.bar(df, x='continent', y='pop', color='continent',
             title='Población por Continente en el Año 2007',
             labels={'pop': 'Población', 'continent': 'Continente'})

# Mostrar el gráfico
fig.show()
````

````python
# Crear el gráfico de barras con Plotly Express
fig = px.bar(df, x='continent', y='lifeExp', color='continent', pattern_shape="continent", pattern_shape_sequence=[".", "x", "+"],
             title='Población por Continente en el Año 2007')

# Mostrar el gráfico
fig.show()
````

````python
df = px.data.gapminder()
df_filtered = df[df['year'].isin([1992, 1997, 2002, 2007])]

fig = px.bar(df_filtered, x='continent', y='pop', color='continent', barmode="group", facet_col="year",
             title='Población por Continente para los años 1992, 1997, 2002 y 2007',
             category_orders={"year": [1992, 1997, 2002, 2007]})

fig.show()
````

````python
fig = px.line(df, x="year", y="lifeExp", color="continent", line_group="country", hover_name="country",
        line_shape="spline", render_mode="svg")
fig.show()
````

````python
import plotly.express as px
df = px.data.gapminder()
fig = px.area(df, x="year", y="pop", color="continent", line_group="country")
fig.show()
````

````python
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
fig.show()
````

````python
df = px.data.gapminder().query("year == 2007")
fig = px.sunburst(df, path=['continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'])
fig.show()
````

##### <span style="color:#4361EE">Coordenadas paralelas</span>

````python
df = px.data.gapminder()
fig = px.parallel_coordinates(df, color="pop", labels={"continent": "Continente",
                      "year": "Año", "pop": "Población",
                      "gdpPercap": "PIB per cápita", "lifeExp": "Esperanza de Vida"},
                    color_continuous_midpoint=df['pop'].median())
fig.show()
````
