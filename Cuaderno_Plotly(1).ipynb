## <span style="color:#4361EE">Introducci√≥n a Plotly</span>

Plotly es una librer√≠a de visualizaci√≥n de datos interactiva y de c√≥digo abierto en Python que permite crear gr√°ficos y visualizaciones de alta calidad. Fue desarrollada para facilitar la tarea de explorar, analizar y presentar datos de manera efectiva. Con Plotly, los cient√≠ficos de datos, analistas y programadores pueden crear visualizaciones atractivas e interactivas con facilidad.

### <span style="color:#4CC9F0"> Principales caracter√≠sticas de Plotly</span>

*Interactividad*: Una de las caracter√≠sticas m√°s destacadas de Plotly es su capacidad para crear gr√°ficos interactivos. Los gr√°ficos generados por Plotly permiten a los usuarios explorar y manipular los datos directamente en el gr√°fico. Puedes hacer zoom, desplazarte, seleccionar puntos espec√≠ficos y obtener informaci√≥n detallada al colocar el cursor sobre los elementos.

*Variedad de gr√°ficos*: Plotly ofrece una amplia variedad de tipos de gr√°ficos, desde gr√°ficos de l√≠neas y gr√°ficos de dispersi√≥n hasta gr√°ficos de barras, histogramas, gr√°ficos de pastel y mapas de calor. Esto permite representar tus datos de diferentes maneras y elegir el tipo de gr√°fico m√°s adecuado para resaltar patrones y relaciones en los datos.

*Facilidad de uso*: Plotly se destaca por su facilidad de uso y sintaxis sencilla. La librer√≠a est√° dise√±ada para ser intuitiva y accesible para aquellos que deseen crear visualizaciones sin una curva de aprendizaje compleja.

*Visualizaci√≥n enriquecida*: Los gr√°ficos generados por Plotly son altamente personalizables. Puedes ajustar aspectos como colores, t√≠tulos, etiquetas de ejes y leyendas para mejorar la presentaci√≥n visual de tus datos.

*Integraci√≥n con entornos de datos*: Plotly se integra perfectamente con Jupyter Notebooks, lo que facilita la creaci√≥n de visualizaciones interactivas junto con el an√°lisis de datos y la narrativa de proyectos.

*Soporte multiplataforma*: Adem√°s de Python, Plotly est√° disponible en otros lenguajes de programaci√≥n como R, JavaScript y Julia, lo que facilita su adopci√≥n en diferentes entornos y flujos de trabajo.

### <span style="color:#4CC9F0">Aplicaciones de Plotly en Ciencia de datos</span>

*Exploraci√≥n de datos*: Plotly es una herramienta poderosa para explorar datos y detectar patrones, tendencias y relaciones en los datos de manera interactiva.

*Presentaci√≥n de resultados*: Con Plotly, puedes crear visualizaciones atractivas y din√°micas para presentar tus resultados y hallazgos de manera clara y convincente.

*Visualizaci√≥n geoespacial*: Plotly permite crear mapas interactivos y mapas cloropl√©ticos para representar datos geoespaciales y resaltar patrones regionales.

*Visualizaci√≥n temporal*: Puedes utilizar Plotly para mostrar cambios en los datos a lo largo del tiempo mediante animaciones y gr√°ficos interactivos.

### <span style="color:#4361EE">Gr√°ficos b√°sicos</span>

#### <span style="color:#4361EE">Scatter</span>

En el primer gr√°fico, utilizamos el conjunto de datos "gapminder" y filtramos los datos para que solo incluyan el a√±o 2007. Luego, creamos un diagrama de dispersi√≥n (scatter plot) utilizando Plotly Express. El eje x representa el PIB per c√°pita, mientras que el eje y representa la esperanza de vida. Los puntos se agrupan por continente y se les asigna un color √∫nico para cada continente. Adem√°s, al pasar el cursor sobre un punto, se mostrar√° el nombre del pa√≠s. Este gr√°fico nos permite visualizar la relaci√≥n entre el PIB per c√°pita, la esperanza de vida y el continente al que pertenece cada pa√≠s en el a√±o 2007.

````python
df = px.data.gapminder().query("year == 2007")
fig = px.scatter(df, x="gdpPercap", y="lifeExp", color="continent", hover_name="country")
fig.show()
````

La interfaz de los gr√°ficos de Pyplot, proporcionada por Plotly, ofrece diversas herramientas que permiten una interacci√≥n din√°mica con las visualizaciones. Entre las funcionalidades m√°s destacadas se encuentran:

1. **Zoom**: Permite acercar o alejar en una parte espec√≠fica del gr√°fico para examinar detalles con mayor precisi√≥n. Puedes seleccionar un √°rea con el rat√≥n para enfocar en esa regi√≥n.

2. **Download Plot as PNG**: Esta opci√≥n te permite descargar el gr√°fico actual como una imagen en formato PNG, lo que facilita su compartici√≥n y uso en otros contextos.

3. **Pan**: Permite desplazar el gr√°fico horizontalmente o verticalmente para visualizar diferentes secciones sin cambiar la escala.

4. **Box Select**: Es una herramienta de selecci√≥n que te permite arrastrar un rect√°ngulo sobre el gr√°fico para resaltar y seleccionar un conjunto espec√≠fico de puntos o datos.

5. **Lasso Select**: Similar a la opci√≥n Box Select, pero en lugar de un rect√°ngulo, puedes dibujar una forma libre (como un lazo) para seleccionar puntos o datos.

6. **Zoom in / Zoom out**: Estas opciones permiten ampliar o reducir el nivel de acercamiento del gr√°fico.

7. **Autoscale**: Ajusta autom√°ticamente el rango de los ejes para que todos los datos sean visibles, evitando recortes y optimizando la visualizaci√≥n.

8. **Reset Axes**: Esta opci√≥n restaura los ejes a su escala original, deshaciendo cualquier modificaci√≥n realizada mediante zoom o pan.

En conjunto, estas herramientas de interacci√≥n proporcionan una experiencia flexible para explorar y analizar gr√°ficos de manera m√°s detallada y adaptada a las necesidades del usuario. La posibilidad de seleccionar √°reas, descargar im√°genes y ajustar el zoom y la escala de los ejes facilitan la comprensi√≥n y el an√°lisis de los datos representados en las visualizaciones generadas con Pyplot.

En el segundo gr√°fico, creamos otro diagrama de dispersi√≥n, pero esta vez ajustamos el tama√±o de los puntos seg√∫n la poblaci√≥n y aplicamos una escala logar√≠tmica al eje x. De esta manera, los pa√≠ses con mayor poblaci√≥n tendr√°n puntos m√°s grandes. Los colores siguen representando los continentes y al pasar el cursor sobre un punto, se mostrar√° el nombre del pa√≠s. Este gr√°fico nos permite comparar la relaci√≥n entre el PIB per c√°pita, la esperanza de vida y la poblaci√≥n de cada pa√≠s en el a√±o 2007, teniendo en cuenta la distribuci√≥n logar√≠tmica del PIB per c√°pita.

````python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent", hover_name="country", log_x=True)
fig.show()
````

En el tercer gr√°fico, creamos un diagrama de dispersi√≥n similar al segundo gr√°fico, pero esta vez agregamos una l√≠nea de tendencia lineal mediante regresi√≥n por m√≠nimos cuadrados. La l√≠nea de tendencia muestra la relaci√≥n general entre el PIB per c√°pita y la esperanza de vida para los pa√≠ses en el a√±o 2007. Los puntos todav√≠a se agrupan por continente y tienen un tama√±o proporcional a la poblaci√≥n. Al pasar el cursor sobre un punto, se mostrar√° el nombre del pa√≠s. La l√≠nea de tendencia "ols" (regresi√≥n por m√≠nimos cuadrados) nos ayuda a identificar la tendencia general de c√≥mo el PIB per c√°pita se relaciona con la esperanza de vida en ese a√±o.

````python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent", hover_name="country", trendline="ols", log_x=True)
fig.show()
````

En el cuarto gr√°fico, agregamos gr√°ficos marginales en los ejes x e y en forma de viol√≠n y caja, respectivamente. Estos gr√°ficos marginales nos permiten explorar la distribuci√≥n de las variables individuales en los ejes x e y mientras observamos el diagrama de dispersi√≥n principal. Adem√°s, mantenemos la l√≠nea de tendencia "ols" para ver la tendencia general entre el PIB per c√°pita y la esperanza de vida. Al pasar el cursor sobre un punto, se mostrar√° el nombre del pa√≠s. Este gr√°fico ofrece una visi√≥n m√°s detallada de c√≥mo se distribuyen las variables y c√≥mo se relacionan entre s√≠ en el a√±o 2007.

````python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent", hover_name="country", marginal_y="violin", marginal_x="box", trendline="ols", log_x=True)
fig.show()
````

En este gr√°fico, utilizamos el conjunto de datos "gapminder" y creamos una matriz de dispersi√≥n (scatter matrix) utilizando Plotly Express. La matriz de dispersi√≥n es una herramienta poderosa para visualizar la relaci√≥n entre m√∫ltiples variables num√©ricas en un solo vistazo.

En la matriz de dispersi√≥n, cada variable num√©rica del conjunto de datos se representa en un eje, y se generan gr√°ficos de dispersi√≥n en los cuadros de la matriz para comparar todas las posibles combinaciones de variables. En este caso, hemos seleccionado las variables "Esperanza de Vida" (lifeExp), "Poblaci√≥n" (pop) y "PIB per c√°pita" (gdpPercap) para crear la matriz de dispersi√≥n.

````python
fig = px.scatter_matrix(df, dimensions=["lifeExp", "pop", "gdpPercap"], color="continent")
fig.show()
````

Creamos un diagrama de dispersi√≥n animado (animated scatter plot) utilizando Plotly Express.

El eje x del gr√°fico representa el PIB per c√°pita, mientras que el eje y representa la esperanza de vida. Los puntos del diagrama de dispersi√≥n se agrupan por continente y se les asigna un color √∫nico para cada continente. Adem√°s, al pasar el cursor sobre un punto, se mostrar√° el nombre del pa√≠s.

Lo m√°s interesante de este gr√°fico es que hemos utilizado el par√°metro animation_frame="year" para animar la visualizaci√≥n a lo largo del tiempo. Cada frame de la animaci√≥n representa un a√±o diferente en el conjunto de datos "gapminder". Esto nos permite visualizar c√≥mo cambian el PIB per c√°pita y la esperanza de vida para diferentes pa√≠ses y continentes a lo largo de los a√±os.

````python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country", facet_col="continent",
           log_x=True, size_max=45, range_x=[100,100000], range_y=[25,90])
fig.show()
````

#### <span style="color:#4361EE">Bar Plot</span>

El eje x representa los continentes, y el eje y representa la poblaci√≥n de cada continente. Cada barra en el gr√°fico representa un continente y su altura est√° determinada por la poblaci√≥n de ese continente en el a√±o 2007. Adem√°s, el color de cada barra se corresponde con el continente que representa, lo que permite identificar f√°cilmente cada grupo.

La leyenda del gr√°fico muestra la relaci√≥n entre los colores y los continentes, lo que facilita la identificaci√≥n de cada grupo. Al pasar el cursor sobre una barra, se mostrar√° el nombre del continente y la poblaci√≥n correspondiente en una etiqueta emergente.

Este gr√°fico de barras nos permite comparar visualmente la poblaci√≥n de cada continente en el a√±o 2007 y nos da una idea de la distribuci√≥n de la poblaci√≥n mundial entre los diferentes continentes. Es una forma efectiva de mostrar la magnitud de la poblaci√≥n de cada regi√≥n y puede ayudarnos a identificar posibles tendencias o disparidades en la distribuci√≥n de la poblaci√≥n en el mundo.

````python
# Crear el gr√°fico de barras con Plotly Express
fig = px.bar(df, x='continent', y='pop', color='continent',
             title='Poblaci√≥n por Continente en el A√±o 2007',
             labels={'pop': 'Poblaci√≥n', 'continent': 'Continente'})

# Mostrar el gr√°fico
fig.show()
````

Se realiza una modificaci√≥n en el uso de los par√°metros pattern_shape y pattern_shape_sequence. Estos par√°metros permiten agregar patrones diferentes en lugar de un color s√≥lido a cada barra del gr√°fico.

````python
# Crear el gr√°fico de barras con Plotly Express
fig = px.bar(df, x='continent', y='lifeExp', color='continent', pattern_shape="continent", pattern_shape_sequence=[".", "x", "+"],
             title='Poblaci√≥n por Continente en el A√±o 2007')

# Mostrar el gr√°fico
fig.show()
````

Este gr√°fico de barras con gr√°fico de mosaico nos permite comparar visualmente la poblaci√≥n de cada continente en los a√±os 1992, 1997, 2002 y 2007. Podemos observar c√≥mo ha variado la poblaci√≥n de cada continente a lo largo de estos a√±os espec√≠ficos y c√≥mo se comparan entre s√≠.

````python
df = px.data.gapminder()
df_filtered = df[df['year'].isin([1992, 1997, 2002, 2007])]

fig = px.bar(df_filtered, x='continent', y='pop', color='continent', barmode="group", facet_col="year",
             title='Poblaci√≥n por Continente para los a√±os 1992, 1997, 2002 y 2007',
             category_orders={"year": [1992, 1997, 2002, 2007]})

fig.show()
````

#### <span style="color:#4361EE">Line Chart</span>

#### <span style="color:#4361EE">Area Chart</span>

El eje x representa los a√±os, mientras que el eje y representa la poblaci√≥n mundial en cada a√±o. Cada √°rea coloreada en el gr√°fico representa un continente y muestra c√≥mo cambia la poblaci√≥n de ese continente a medida que avanzan los a√±os.

Este gr√°fico de √°rea nos permite observar c√≥mo ha evolucionado la poblaci√≥n mundial a lo largo del tiempo y c√≥mo han contribuido diferentes continentes a ese crecimiento. Tambi√©n podemos identificar per√≠odos de crecimiento o disminuci√≥n de la poblaci√≥n para cada continente y comparar su participaci√≥n relativa en diferentes momentos hist√≥ricos.

#### <span style="color:#4361EE">timeline/Gantt charts</span>

El gr√°fico de l√≠neas de tiempo nos permite visualizar de manera clara y concisa la programaci√≥n de tareas para diferentes recursos en un per√≠odo de tiempo determinado. Podemos observar la superposici√≥n de tareas, duraci√≥n y distribuci√≥n de las tareas asignadas a cada recurso.

En este gr√°fico de l√≠neas de tiempo, utilizamos un conjunto de datos que contiene informaci√≥n sobre tareas y su duraci√≥n asignada a diferentes recursos (personas). El conjunto de datos tiene tres filas, cada una representando una tarea espec√≠fica junto con su fecha de inicio, fecha de finalizaci√≥n y el recurso (persona) asignado a esa tarea.

````python
df = pd.DataFrame([
    dict(Task="Job A", Start='2009-01-01', Finish='2009-02-28', Resource="Alex"),
    dict(Task="Job B", Start='2009-03-05', Finish='2009-04-15', Resource="Alex"),
    dict(Task="Job C", Start='2009-02-20', Finish='2009-05-30', Resource="Max")
])

fig = px.timeline(df, x_start="Start", x_end="Finish", y="Resource", color="Resource")
fig.show()
````

#### <span style="color:#4361EE">funnel charts</span>

Este gr√°fico de embudo nos permite visualizar de manera clara y efectiva c√≥mo se reduce el n√∫mero de elementos a medida que avanzan por las diferentes etapas del proceso. Es especialmente √∫til para mostrar el rendimiento en cada etapa del proceso y c√≥mo se reduce el n√∫mero de elementos a medida que se avanza hacia la etapa final.

En este gr√°fico de embudo, estamos visualizando un proceso en el que hay diferentes etapas y el n√∫mero de elementos que pasan por cada etapa. Para crear este gr√°fico, utilizamos la funci√≥n px.funnel() de Plotly Express.

##### <span style="color:#4361EE">Empirical Cumulative Distribution Function (ECDF) charts.</span>

Este gr√°fico ECDF nos permite visualizar c√≥mo se distribuye la esperanza de vida para cada continente y c√≥mo se comparan entre s√≠. Tambi√©n nos permite observar si las distribuciones son similares o si existen diferencias significativas entre los continentes en t√©rminos de esperanza de vida.

La funci√≥n de distribuci√≥n emp√≠rica acumulada (ECDF) es una funci√≥n que muestra la probabilidad acumulada de que una variable tome un valor menor o igual a un valor espec√≠fico. En este caso, el eje x representa los valores de la esperanza de vida, y el eje y representa la probabilidad acumulada.

El gr√°fico muestra una l√≠nea ECDF para cada continente, cada una con un color diferente para distinguirlas. La l√≠nea ECDF de cada continente se construye al ordenar los valores de la esperanza de vida para ese continente y luego calcular la fracci√≥n de datos que tienen un valor menor o igual al valor espec√≠fico en el eje x. A medida que avanzamos en el eje x, la l√≠nea ECDF aumenta gradualmente y finalmente alcanza 1 en el valor m√°ximo de esperanza de vida, lo que indica que todos los datos est√°n por debajo de ese valor.

````python
df = px.data.gapminder()
fig = px.ecdf(df, x="lifeExp", color="continent")
fig.show()
````
