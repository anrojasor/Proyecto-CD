## <span style="color:#4361EE">Introducción a Plotly</span>

Plotly es una librería de visualización de datos interactiva y de código abierto en Python que permite crear gráficos y visualizaciones de alta calidad. Fue desarrollada para facilitar la tarea de explorar, analizar y presentar datos de manera efectiva. Con Plotly, los científicos de datos, analistas y programadores pueden crear visualizaciones atractivas e interactivas con facilidad.

### <span style="color:#4CC9F0"> Principales características de Plotly</span>

*Interactividad*: Una de las características más destacadas de Plotly es su capacidad para crear gráficos interactivos. Los gráficos generados por Plotly permiten a los usuarios explorar y manipular los datos directamente en el gráfico. Puedes hacer zoom, desplazarte, seleccionar puntos específicos y obtener información detallada al colocar el cursor sobre los elementos.

*Variedad de gráficos*: Plotly ofrece una amplia variedad de tipos de gráficos, desde gráficos de líneas y gráficos de dispersión hasta gráficos de barras, histogramas, gráficos de pastel y mapas de calor. Esto permite representar tus datos de diferentes maneras y elegir el tipo de gráfico más adecuado para resaltar patrones y relaciones en los datos.

*Facilidad de uso*: Plotly se destaca por su facilidad de uso y sintaxis sencilla. La librería está diseñada para ser intuitiva y accesible para aquellos que deseen crear visualizaciones sin una curva de aprendizaje compleja.

*Visualización enriquecida*: Los gráficos generados por Plotly son altamente personalizables. Puedes ajustar aspectos como colores, títulos, etiquetas de ejes y leyendas para mejorar la presentación visual de tus datos.

*Integración con entornos de datos*: Plotly se integra perfectamente con Jupyter Notebooks, lo que facilita la creación de visualizaciones interactivas junto con el análisis de datos y la narrativa de proyectos.

*Soporte multiplataforma*: Además de Python, Plotly está disponible en otros lenguajes de programación como R, JavaScript y Julia, lo que facilita su adopción en diferentes entornos y flujos de trabajo.

### <span style="color:#4CC9F0">Aplicaciones de Plotly en Ciencia de datos</span>

*Exploración de datos*: Plotly es una herramienta poderosa para explorar datos y detectar patrones, tendencias y relaciones en los datos de manera interactiva.

*Presentación de resultados*: Con Plotly, puedes crear visualizaciones atractivas y dinámicas para presentar tus resultados y hallazgos de manera clara y convincente.

*Visualización geoespacial*: Plotly permite crear mapas interactivos y mapas cloropléticos para representar datos geoespaciales y resaltar patrones regionales.

*Visualización temporal*: Puedes utilizar Plotly para mostrar cambios en los datos a lo largo del tiempo mediante animaciones y gráficos interactivos.

### <span style="color:#4361EE">Gráficos básicos</span>

#### <span style="color:#4361EE">Scatter</span>

En el primer gráfico, utilizamos el conjunto de datos "gapminder" y filtramos los datos para que solo incluyan el año 2007. Luego, creamos un diagrama de dispersión (scatter plot) utilizando Plotly Express. El eje x representa el PIB per cápita, mientras que el eje y representa la esperanza de vida. Los puntos se agrupan por continente y se les asigna un color único para cada continente. Además, al pasar el cursor sobre un punto, se mostrará el nombre del país. Este gráfico nos permite visualizar la relación entre el PIB per cápita, la esperanza de vida y el continente al que pertenece cada país en el año 2007.

````python
df = px.data.gapminder().query("year == 2007")
fig = px.scatter(df, x="gdpPercap", y="lifeExp", color="continent", hover_name="country")
fig.show()
````

La interfaz de los gráficos de Pyplot, proporcionada por Plotly, ofrece diversas herramientas que permiten una interacción dinámica con las visualizaciones. Entre las funcionalidades más destacadas se encuentran:

1. **Zoom**: Permite acercar o alejar en una parte específica del gráfico para examinar detalles con mayor precisión. Puedes seleccionar un área con el ratón para enfocar en esa región.

2. **Download Plot as PNG**: Esta opción te permite descargar el gráfico actual como una imagen en formato PNG, lo que facilita su compartición y uso en otros contextos.

3. **Pan**: Permite desplazar el gráfico horizontalmente o verticalmente para visualizar diferentes secciones sin cambiar la escala.

4. **Box Select**: Es una herramienta de selección que te permite arrastrar un rectángulo sobre el gráfico para resaltar y seleccionar un conjunto específico de puntos o datos.

5. **Lasso Select**: Similar a la opción Box Select, pero en lugar de un rectángulo, puedes dibujar una forma libre (como un lazo) para seleccionar puntos o datos.

6. **Zoom in / Zoom out**: Estas opciones permiten ampliar o reducir el nivel de acercamiento del gráfico.

7. **Autoscale**: Ajusta automáticamente el rango de los ejes para que todos los datos sean visibles, evitando recortes y optimizando la visualización.

8. **Reset Axes**: Esta opción restaura los ejes a su escala original, deshaciendo cualquier modificación realizada mediante zoom o pan.

En conjunto, estas herramientas de interacción proporcionan una experiencia flexible para explorar y analizar gráficos de manera más detallada y adaptada a las necesidades del usuario. La posibilidad de seleccionar áreas, descargar imágenes y ajustar el zoom y la escala de los ejes facilitan la comprensión y el análisis de los datos representados en las visualizaciones generadas con Pyplot.

En el segundo gráfico, creamos otro diagrama de dispersión, pero esta vez ajustamos el tamaño de los puntos según la población y aplicamos una escala logarítmica al eje x. De esta manera, los países con mayor población tendrán puntos más grandes. Los colores siguen representando los continentes y al pasar el cursor sobre un punto, se mostrará el nombre del país. Este gráfico nos permite comparar la relación entre el PIB per cápita, la esperanza de vida y la población de cada país en el año 2007, teniendo en cuenta la distribución logarítmica del PIB per cápita.

````python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent", hover_name="country", log_x=True)
fig.show()
````

En el tercer gráfico, creamos un diagrama de dispersión similar al segundo gráfico, pero esta vez agregamos una línea de tendencia lineal mediante regresión por mínimos cuadrados. La línea de tendencia muestra la relación general entre el PIB per cápita y la esperanza de vida para los países en el año 2007. Los puntos todavía se agrupan por continente y tienen un tamaño proporcional a la población. Al pasar el cursor sobre un punto, se mostrará el nombre del país. La línea de tendencia "ols" (regresión por mínimos cuadrados) nos ayuda a identificar la tendencia general de cómo el PIB per cápita se relaciona con la esperanza de vida en ese año.

````python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent", hover_name="country", trendline="ols", log_x=True)
fig.show()
````

En el cuarto gráfico, agregamos gráficos marginales en los ejes x e y en forma de violín y caja, respectivamente. Estos gráficos marginales nos permiten explorar la distribución de las variables individuales en los ejes x e y mientras observamos el diagrama de dispersión principal. Además, mantenemos la línea de tendencia "ols" para ver la tendencia general entre el PIB per cápita y la esperanza de vida. Al pasar el cursor sobre un punto, se mostrará el nombre del país. Este gráfico ofrece una visión más detallada de cómo se distribuyen las variables y cómo se relacionan entre sí en el año 2007.

````python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent", hover_name="country", marginal_y="violin", marginal_x="box", trendline="ols", log_x=True)
fig.show()
````

En este gráfico, utilizamos el conjunto de datos "gapminder" y creamos una matriz de dispersión (scatter matrix) utilizando Plotly Express. La matriz de dispersión es una herramienta poderosa para visualizar la relación entre múltiples variables numéricas en un solo vistazo.

En la matriz de dispersión, cada variable numérica del conjunto de datos se representa en un eje, y se generan gráficos de dispersión en los cuadros de la matriz para comparar todas las posibles combinaciones de variables. En este caso, hemos seleccionado las variables "Esperanza de Vida" (lifeExp), "Población" (pop) y "PIB per cápita" (gdpPercap) para crear la matriz de dispersión.

````python
fig = px.scatter_matrix(df, dimensions=["lifeExp", "pop", "gdpPercap"], color="continent")
fig.show()
````

Creamos un diagrama de dispersión animado (animated scatter plot) utilizando Plotly Express.

El eje x del gráfico representa el PIB per cápita, mientras que el eje y representa la esperanza de vida. Los puntos del diagrama de dispersión se agrupan por continente y se les asigna un color único para cada continente. Además, al pasar el cursor sobre un punto, se mostrará el nombre del país.

Lo más interesante de este gráfico es que hemos utilizado el parámetro animation_frame="year" para animar la visualización a lo largo del tiempo. Cada frame de la animación representa un año diferente en el conjunto de datos "gapminder". Esto nos permite visualizar cómo cambian el PIB per cápita y la esperanza de vida para diferentes países y continentes a lo largo de los años.

````python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country", facet_col="continent",
           log_x=True, size_max=45, range_x=[100,100000], range_y=[25,90])
fig.show()
````

#### <span style="color:#4361EE">Bar Plot</span>

El eje x representa los continentes, y el eje y representa la población de cada continente. Cada barra en el gráfico representa un continente y su altura está determinada por la población de ese continente en el año 2007. Además, el color de cada barra se corresponde con el continente que representa, lo que permite identificar fácilmente cada grupo.

La leyenda del gráfico muestra la relación entre los colores y los continentes, lo que facilita la identificación de cada grupo. Al pasar el cursor sobre una barra, se mostrará el nombre del continente y la población correspondiente en una etiqueta emergente.

Este gráfico de barras nos permite comparar visualmente la población de cada continente en el año 2007 y nos da una idea de la distribución de la población mundial entre los diferentes continentes. Es una forma efectiva de mostrar la magnitud de la población de cada región y puede ayudarnos a identificar posibles tendencias o disparidades en la distribución de la población en el mundo.

````python
# Crear el gráfico de barras con Plotly Express
fig = px.bar(df, x='continent', y='pop', color='continent',
             title='Población por Continente en el Año 2007',
             labels={'pop': 'Población', 'continent': 'Continente'})

# Mostrar el gráfico
fig.show()
````

Se realiza una modificación en el uso de los parámetros pattern_shape y pattern_shape_sequence. Estos parámetros permiten agregar patrones diferentes en lugar de un color sólido a cada barra del gráfico.

````python
# Crear el gráfico de barras con Plotly Express
fig = px.bar(df, x='continent', y='lifeExp', color='continent', pattern_shape="continent", pattern_shape_sequence=[".", "x", "+"],
             title='Población por Continente en el Año 2007')

# Mostrar el gráfico
fig.show()
````

Este gráfico de barras con gráfico de mosaico nos permite comparar visualmente la población de cada continente en los años 1992, 1997, 2002 y 2007. Podemos observar cómo ha variado la población de cada continente a lo largo de estos años específicos y cómo se comparan entre sí.

````python
df = px.data.gapminder()
df_filtered = df[df['year'].isin([1992, 1997, 2002, 2007])]

fig = px.bar(df_filtered, x='continent', y='pop', color='continent', barmode="group", facet_col="year",
             title='Población por Continente para los años 1992, 1997, 2002 y 2007',
             category_orders={"year": [1992, 1997, 2002, 2007]})

fig.show()
````

#### <span style="color:#4361EE">Area Chart</span>

El eje x representa los años, mientras que el eje y representa la población mundial en cada año. Cada área coloreada en el gráfico representa un continente y muestra cómo cambia la población de ese continente a medida que avanzan los años.

También podemos identificar períodos de crecimiento o disminución de la población para cada continente y comparar su participación relativa en diferentes momentos históricos.

#### <span style="color:#4361EE">timeline/Gantt charts</span>

El gráfico de líneas de tiempo nos permite visualizar de manera clara y concisa la programación de tareas para diferentes recursos en un período de tiempo determinado. Podemos observar la superposición de tareas, duración y distribución de las tareas asignadas a cada recurso.

En este gráfico de líneas de tiempo, utilizamos un conjunto de datos que contiene información sobre tareas y su duración asignada a diferentes recursos (personas). El conjunto de datos tiene tres filas, cada una representando una tarea específica junto con su fecha de inicio, fecha de finalización y el recurso (persona) asignado a esa tarea.

````python
df = pd.DataFrame([
    dict(Task="Job A", Start='2009-01-01', Finish='2009-02-28', Resource="Alex"),
    dict(Task="Job B", Start='2009-03-05', Finish='2009-04-15', Resource="Alex"),
    dict(Task="Job C", Start='2009-02-20', Finish='2009-05-30', Resource="Max")
])

fig = px.timeline(df, x_start="Start", x_end="Finish", y="Resource", color="Resource")
fig.show()
````

#### <span style="color:#4361EE">funnel charts</span>

Este gráfico de embudo nos permite visualizar de manera clara y efectiva cómo se reduce el número de elementos a medida que avanzan por las diferentes etapas del proceso. Es especialmente útil para mostrar el rendimiento en cada etapa del proceso y cómo se reduce el número de elementos a medida que se avanza hacia la etapa final.

En este gráfico de embudo, estamos visualizando un proceso en el que hay diferentes etapas y el número de elementos que pasan por cada etapa. Para crear este gráfico, utilizamos la función px.funnel() de Plotly Express.

````python
data = dict(
    number=[39, 27.4, 20.6, 11, 2],
    stage=["Website visit", "Downloads", "Potential customers", "Requested price", "Invoice sent"])
fig = px.funnel(data, x='number', y='stage')
fig.show()
````

#### <span style="color:#4361EE">Gráficos de Torta</span>

El gráfico de pastel representa la población de los países europeos en el año 2007. Cada porción del pastel corresponde a un país y su tamaño está determinado por la población de ese país. Los países con mayor población tendrán una porción más grande del pastel, mientras que los países con población menor a 2 millones se agruparán bajo la etiqueta 'Other countries'.

````python
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
fig.show()
````

El gráfico de Sunburst es una representación visual jerárquica que muestra la estructura de los datos en forma de anillos concéntricos. Cada anillo representa una categoría y se divide en segmentos que representan subcategorías. En este caso, hemos configurado el gráfico para mostrar la población de cada país agrupada por continente y país.

````python
df = px.data.gapminder().query("year == 2007")
fig = px.sunburst(df, path=['continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'])
fig.show()
````

El gráfico de treemap es una forma de visualización jerárquica que muestra la relación entre categorías anidadas mediante rectángulos. En este caso, las categorías anidadas son "world" (mundo) -> "continent" (continente) -> "country" (país). Cada rectángulo en el gráfico representa un país y su tamaño está determinado por la población de ese país en el año 2007.

````python
df = px.data.gapminder().query("year == 2007")
fig = px.treemap(df, path=[px.Constant('world'), 'continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'])
fig.show()
````

El gráfico icicle es una representación jerárquica y jerárquica compacta que muestra la estructura de los datos en una estructura de árbol. Cada rectángulo en el gráfico representa una región y está coloreado según la esperanza de vida promedio en ese país. La altura de cada rectángulo corresponde a la población de esa región.

````python
df = px.data.gapminder().query("year == 2007")
fig = px.icicle(df, path=[px.Constant('world'), 'continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'])
fig.show()
````

#### <span style="color:#4361EE">Gráficos de Distribución</span>

##### <span style="color:#4361EE">Histograma</span>

El eje x representa los valores de la esperanza de vida (lifeExp), y el eje y muestra la frecuencia con la que aparecen estos valores en el conjunto de datos. Cada barra en el histograma representa un rango de valores de esperanza de vida y muestra cuántos países tienen valores de esperanza de vida dentro de ese rango.

En general, este gráfico de histograma nos permite visualizar cómo se distribuye la esperanza de vida en diferentes continentes y cómo varía dentro de cada continente. También podemos identificar si hay diferencias significativas en la distribución de la esperanza de vida entre los continentes o si hay algún rango de valores de esperanza de vida común a varios continentes.

El parámetro marginal="rug" agrega un marcador tipo "rug" a lo largo del eje x. Los marcadores rug son pequeñas líneas verticales que indican la ubicación de cada dato en el eje x. Esto nos ayuda a ver la densidad de datos y la distribución de la esperanza de vida en el conjunto de datos.

##### <span style="color:#4361EE">Empirical Cumulative Distribution Function (ECDF) charts.</span>

Este gráfico ECDF nos permite visualizar cómo se distribuye la esperanza de vida para cada continente y cómo se comparan entre sí. También nos permite observar si las distribuciones son similares o si existen diferencias significativas entre los continentes en términos de esperanza de vida.

La función de distribución empírica acumulada (ECDF) es una función que muestra la probabilidad acumulada de que una variable tome un valor menor o igual a un valor específico. En este caso, el eje x representa los valores de la esperanza de vida, y el eje y representa la probabilidad acumulada.

El gráfico muestra una línea ECDF para cada continente, cada una con un color diferente para distinguirlas. La línea ECDF de cada continente se construye al ordenar los valores de la esperanza de vida para ese continente y luego calcular la fracción de datos que tienen un valor menor o igual al valor específico en el eje x. A medida que avanzamos en el eje x, la línea ECDF aumenta gradualmente y finalmente alcanza 1 en el valor máximo de esperanza de vida, lo que indica que todos los datos están por debajo de ese valor.

````python
df = px.data.gapminder()
fig = px.ecdf(df, x="lifeExp", color="continent")
fig.show()
````
