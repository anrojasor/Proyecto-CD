# Aprendizaje Estadístico

El aprendizaje estadístico, también conocido como aprendizaje automático (machine learning en inglés), es una rama de la inteligencia artificial y la estadística que se enfoca en desarrollar algoritmos y técnicas que permitan a las computadoras aprender patrones, tendencias y relaciones a partir de datos sin ser programadas explícitamente para cada tarea específica.

En lugar de seguir instrucciones específicas, los algoritmos de aprendizaje estadístico utilizan datos de entrenamiento para ajustar sus parámetros y desarrollar modelos que puedan hacer predicciones o tomar decisiones basadas en nuevos datos. La idea central es que la computadora aprenda de los ejemplos y encuentre patrones en los datos por sí misma.

Existen diferentes tipos de aprendizaje estadístico, los principales son:

1. **Aprendizaje Supervisado:** En este enfoque, el algoritmo se entrena utilizando un conjunto de datos etiquetado, es decir, un conjunto de datos donde ya se conoce la respuesta correcta para cada ejemplo. El objetivo es que el modelo aprenda a mapear las entradas a las salidas correctas y pueda hacer predicciones precisas para nuevas instancias no etiquetadas.

2. **Aprendizaje No Supervisado:** En este caso, el algoritmo se entrena utilizando un conjunto de datos no etiquetado. El objetivo es que el modelo encuentre patrones y estructuras ocultas en los datos sin guía explícita sobre las salidas esperadas. El agrupamiento (clustering) y la reducción de dimensionalidad son ejemplos de técnicas de aprendizaje no supervisado.

3. **Aprendizaje por Refuerzo:** En este enfoque, el algoritmo aprende a través de la interacción con un ambiente o sistema. El algoritmo toma acciones en el ambiente y recibe recompensas o penalizaciones en función de sus acciones. El objetivo es aprender una política óptima que maximice la recompensa a lo largo del tiempo.

El aprendizaje estadístico tiene una amplia gama de aplicaciones en diversos campos, como reconocimiento de voz, visión por computadora, procesamiento de lenguaje natural, sistemas de recomendación, análisis financiero, diagnóstico médico, pronóstico del clima, entre muchos otros. Es una poderosa herramienta que ha demostrado su capacidad para resolver problemas complejos y tomar decisiones precisas basadas en datos de gran tamaño y complejidad.

Un ejemplo de modelo utilizado en aprendizaje estadístico es la Regresión Lineal, que es una técnica para modelar la relación entre una variable dependiente (objetivo) y una o más variables independientes (características). La regresión lineal se puede representar mediante la siguiente fórmula matemática:

En el caso más simple, con una sola variable independiente (x) y una variable dependiente (y), la regresión lineal se expresa como:

\[ $y = \beta_0 + \beta_1 \cdot x + \varepsilon$ \]

Donde:
- \(y\) es la variable dependiente (objetivo) que queremos predecir.
- \(x\) es la variable independiente (característica) que utilizamos para hacer la predicción.
- \($\beta_0$\) es la ordenada al origen o el término de intersección.
- \($\beta_1$\) es la pendiente o coeficiente de regresión que representa el cambio en \(y\) por un cambio unitario en \(x\).
- \($\varepsilon$\) es el término de error, que representa la diferencia entre el valor observado y el valor predicho por el modelo.

El objetivo del aprendizaje estadístico es encontrar los valores óptimos para los coeficientes \($\beta_0$\) y \($\beta_1$\) que minimicen el error de predicción en el conjunto de datos de entrenamiento, lo que permitirá que el modelo generalice bien a nuevos datos no vistos.

# Ejemplo

Importamos las librerías y creamos un conjunto de datos para el ejemplo

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo (área en pies cuadrados y precio de las casas)
data = {'Area (pies cuadrados)': [1000, 1500, 2000, 2500, 3000],
        'Precio (USD)': [300000, 450000, 500000, 550000, 600000]}
df = pd.DataFrame(data)
```

A continuación, visualizamps los datos para entender la relación entre el área y el precio:

```python
plt.scatter(df['Area (pies cuadrados)'], df['Precio (USD)'])
plt.xlabel('Área (pies cuadrados)')
plt.ylabel('Precio (USD)')
plt.title('Relación entre el Área y el Precio de las Casas')
plt.show()
```

Ahora, preparamos los datos para el entrenamiento del modelo:

```python
X = df[['Area (pies cuadrados)']]
y = df['Precio (USD)']
```

Creamos y entrenamos el modelo de regresión lineal:

```python
model = LinearRegression()
model.fit(X, y)
```

Realizamos predicciones con el modelo entrenado:

```python
area_nueva = 1800
precio_predicho = model.predict([[area_nueva]])
print(f'El precio estimado para una casa de {area_nueva} pies cuadrados es de ${precio_predicho[0]:,.2f}')
```

Visualizamos la línea de regresión en el gráfico:

```python
plt.scatter(df['Area (pies cuadrados)'], df['Precio (USD)'])
plt.plot(df['Area (pies cuadrados)'], model.predict(X), color='red')
plt.xlabel('Área (pies cuadrados)')
plt.ylabel('Precio (USD)')
plt.title('Relación entre el Área y el Precio de las Casas con Regresión Lineal')
plt.show()
```

Con este ejemplo, hemos utilizado la regresión lineal para ajustar una línea que mejor se ajusta a los datos y, a partir de esta línea, hemos realizado predicciones para el precio de una casa con una determinada área en pies cuadrados. El aprendizaje estadístico nos permite construir modelos que pueden ser útiles para hacer predicciones y tomar decisiones basadas en datos históricos.
